### Genre mapping to the right albums

```python
album_genre_map = {
    "1989 (Taylor's Version)": "Pop",
    "After Hours": "R&B",
    "Austin": "Pop Rock",
    "Autumn Variations": "Pop",
    "Happier Than Ever": "Alternative/Indie",
    "Nadie Sabe Lo Que Va a Pasar Mañana": "Latin Trap",
    "Scarlet": "Rap",
    "Eternal Sunshine": "Pop",
    "Future Nostalgia": "Dance Pop",
    "BORN PINK": "K-pop",
    "SOS": "R&B",
    "MAÑANA SERÁ BONITO": "Reggaeton",
    "Proof": "K-pop",
    "For All The Dogs": "Hip-Hop",
    "Guts": "Pop Rock"
}
```

```python
spotify_data["Genre_Cleaned"] = spotify_data["Album"].map(album_genre_map)
```

```python
unique_artist_album_genre = spotify_data[["Artist", "Album", "Genre_Cleaned"]].drop_duplicates().sort_values(by="Artist")
unique_artist_album_genre 
```
